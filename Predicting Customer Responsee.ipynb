### Now we will create a Regression Model



 Importing Spark SQL and Spark ML Libraries

```python
%scala

import org.apache.spark.sql.functions._
import org.apache.spark.sql.Row
import org.apache.spark.sql.types._

import org.apache.spark.ml.classification.LogisticRegression
import org.apache.spark.ml.feature.VectorAssembler
```

### Preparing data for Training
For training the regression model we will require training dataset. This dataset should consist of a vector which are the numeric values of features and a lable column. In Here we are using  **VectorAssembler** class which will transform the feature columns into a vector and also will rename the  **Y** column to **label**.

```python
%scala

var StringfeatureCol = Array("job", "marital", "education", "default", "housing", "loan", "contact", "month", "day_of_week", "poutcome", "y")
```

### Pipeline
The predictive model will require multiple stages of feasture preparation. 

A pipeline is a series of *transformer* and *estimator* stages that generates a DataFrame which is used for modelling and also to train the predictive model. 
Here the pipeline consists of two stages.

- A **StringIndexer** estimator which will convert all the string values to indexes which will be used for categorical features.
- A **VectorAssembler** will convert all the categorical features into a single vector.

```python
%scala

import org.apache.spark.ml.attribute.Attribute
import org.apache.spark.ml.feature.{IndexToString, StringIndexer}
import org.apache.spark.ml.{Pipeline, PipelineModel}

val indexers = StringfeatureCol.map { colName =>
  new StringIndexer().setInputCol(colName).setOutputCol(colName + "_indexed")
}

val pipeline = new Pipeline()
                    .setStages(indexers)      

val bankingDF = pipeline.fit(bankDF).transform(bankDF)
```

```python
%scala

bankingDF.printSchema()
```

```python
%scala

bankingDF.show()
```

### Split the Data
Here we will be splitting the source data into training data and testing data. We are using 70% of the source data for training and 30% for training. In the testing data, the **label** column is renamed to **trueLabel** so we can use it later to compare predicted labels with known actual values.

```python
%scala

val splits = bankingDF.randomSplit(Array(0.7, 0.3))
val train = splits(0)
val test = splits(1)
val train_rows = train.count()
val test_rows = test.count()
println("Training Rows: " + train_rows + " Testing Rows: " + test_rows)
```

```python
%scala

val assembler = new VectorAssembler().setInputCols(Array("age", "duration", "campaign", "pdays", "previous", "empvarrate", "conspriceidx", "consconfidx", "euribor3m", 
"nremployed", "job_indexed", "marital_indexed", "education_indexed", "default_indexed", "housing_indexed", "loan_indexed", "contact_indexed", "month_indexed", "day_of_week_indexed", "poutcome_indexed")).setOutputCol("features")

val training = assembler.transform(train).select($"features", $"y_indexed".alias("label"))

training.show()
```

### Training the Regression Model
Here the regression model will be trained using the training data. This can be done by creating an instance of the regression algorithm we want to use and then use the **fit** method to train the model based on the training dataframe. We will be using *Logistic Regression* algorithm which is supported in the spark.ml API.

```python
%scala
import org.apache.spark.ml.classification.LogisticRegression

val lr = new LogisticRegression().setLabelCol("label").setFeaturesCol("features").setMaxIter(10).setRegParam(0.3)
val model = lr.fit(training)
println("Model Trained!")
```

### Prepare the Testing Data
Now as the model is trained, it can be tested on the testing data which is reserved for testing previoulsy. Firstly the features of  testing data will need to be transformed into a vector. Here the   **y_indexed** column will be reamned to **trueLabel**.

```python
%scala

val testing = assembler.transform(test).select($"features", $"y_indexed".alias("trueLabel"))
testing.show()
```

### Test the Model
Now we will use the **transform** method of the model to generate some predictions. We can compare the known true value from testing data to the predicted value.

```python
%scala

val prediction = model.transform(testing)
val predicted = prediction.select("features", "prediction", "trueLabel")
predicted.show()
```

### Model Evaluation

Spark.mllib consists of many algorithms that can bes used for predictions on the data. After using these algorithms , it is mandatory to evaluate the performance of the model. We will use the API provided by spark.mllib to evaluate our model.

```python
%scala
import org.apache.spark.ml.evaluation.BinaryClassificationEvaluator

val evaluator = new BinaryClassificationEvaluator().setLabelCol("trueLabel").setRawPredictionCol("rawPrediction").setMetricName("areaUnderROC")
val auc = evaluator.evaluate(prediction)
println("AUC = " + (auc))
```

### Train a Classification Model (Decision tree classifier)
Now we will train our second model on the training data.This can be done by creating an instance of the Decision tree classifier algorithm we want to use and then use the **fit** method to train the model based on the training dataframe. We will be using *Decision tree classifier algorithm*  which is supported in the spark.ml API.

```python
%scala

import org.apache.spark.ml.classification.DecisionTreeClassificationModel
import org.apache.spark.ml.classification.DecisionTreeClassifier
import org.apache.spark.ml.evaluation.MulticlassClassificationEvaluator

val dt = new DecisionTreeClassifier().setLabelCol("label").setFeaturesCol("features")

val model = dt.fit(training)

println("Model Trained!")
```

### Test the Model
Now we will use the **transform** method of the model to generate some predictions. We can compare the known true value from testing data to the predicted value.

```python
%scala

val prediction = model.transform(testing)
val predicted = prediction.select("features", "prediction", "trueLabel")
predicted.show()
```

### Model Evaluation

Spark.mllib consists of many algorithms that can bes used for predictions on the data. After using these algorithms , it is mandatory to evaluate the performance of the model. We will use the API provided by spark.mllib to evaluate our model.

```python
%scala

val evaluator = new MulticlassClassificationEvaluator()
  .setLabelCol("trueLabel")
  .setPredictionCol("prediction")
  .setMetricName("accuracy")
val accuracy = evaluator.evaluate(prediction)
```
